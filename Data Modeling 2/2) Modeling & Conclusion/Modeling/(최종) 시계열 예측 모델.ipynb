```python
df = pd.read_csv('/content/drive/Shareddrives/2021 팜 경진대회/3 코드/3 버스노선 관련/예측 돌린 결과/1.csv')
df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1, inplace=True)

data = df[['202012', '202101', '202102', '202103',
       '202104', '202105', '202106', '202107', '202108', '202109', '202110',
       '202111']]
```

```python
from fbprophet import Prophet
what_i_need = []
forecast_data_list = []
for i in range(len(data)):

  row = pd.DataFrame(data.iloc[i])
  row.reset_index(inplace=True)
  row.columns = ['ds', 'y']
  row['ds'] = pd.to_datetime(row['ds'], format="%Y%m")

  row['cap'] = row['y'].max() * 1.5
  row['floor'] =0
  try:
    prophet = Prophet(growth = 'logistic',
                  changepoint_prior_scale=0.5)
    prophet.fit(row) 

    future_data = prophet.make_future_dataframe(periods=3, freq='m')
    future_data['cap'] = row['y'].max() * 1.5
    future_data['floor'] = 0
    forecast_data = prophet.predict(future_data)
    forecast_data_list.append(forecast_data)
    
    need = forecast_data[['ds', 'yhat']].tail(3)
  except:
    need = pd.DataFrame()
  what_i_need.append(need)
  
  print(i,"th row Completed")
```

```python
col = ['2021-11-30', '2021-12-31', '2022-01-31']
predicted_list = []
for i in range(len(what_i_need)):
  now = what_i_need[i]
  now_list = []
  for j in range(3):
    now_value = now.iloc[j, 1]
    now_list.append(now_value)
  predicted_list.append(now_list)
predicted_df = pd.DataFrame(predicted_list, columns=col)

result = pd.concat([df, predicted_df], axis=1)
```
