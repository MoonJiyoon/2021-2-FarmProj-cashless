```python
busstop = pd.read_csv('/content/drive/Shareddrives/2021 팜 경진대회/3 코드/3 버스노선 관련/apicall_all.csv', encoding='cp949')
busstop = busstop.drop('Unnamed: 0', axis=1)
busstop
```

```python
original = pd.read_csv('/content/drive/Shareddrives/2021 팜 경진대회/2 데이터/0 가공완료 데이터/서울시 버스정류장 중복 제거 _통합.csv', encoding='cp949')
original = original[['정류소명', '행정동']]
original
```

```python
# 불러온 api와 api에 넣은 input의 행정동 매핑
busstop['행정동매핑'] = ''
for idi, i in enumerate(original['정류소명']):
  for idj, j in enumerate(busstop['stnm']):
    if j==i:
      busstop.iloc[idj,-1] = original.iloc[idi,1]
# busstop['stid'] = busstop['stid'].astype('int')
busstop
```

```python
busstop = busstop[['arsid', 'stid', 'stnm', '행정동매핑', 'x', 'y']]
busstop.columns = ['arsid', '정류장ID', '정류소명', '행정동매핑', 'x', 'y']
busstop.to_csv('버스정류장 - 행정동 매핑.csv', encoding='cp949')
```
