```python
geo_path = 'skorea_submunicipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding = 'utf-8'))
geo_str
```

```python
dongData = pd.read_excel('산일구비_최종_데이터셋(코드매핑).xlsx')
for i in range(len(dongData)) :
    dongData.loc[i,"행정동코드"] = str(dongData.loc[i,"행정동코드"])
dongData
```

```python
seoul_map = folium.Map(location=[37.5502, 126.982], zoom_start=10.5, tiles='cartodbpositron')
seoul_map.choropleth(geo_data=geo_str,
                    data=dongData,
                    columns=['행정동코드','65세 이상 인구 수'],
                    fill_color='PuRd',
                    key_on='properties.code',
                    highlight=True,
                    fill_opacity=0.5, line_opacity=1,
                    legend_name='number')
```
