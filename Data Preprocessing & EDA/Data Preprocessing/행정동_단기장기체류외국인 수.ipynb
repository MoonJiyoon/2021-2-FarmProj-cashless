```python
import pandas as pd
```

```python
df = pd.read_csv('행정동별 서울생활인구(단기체류 외국인).csv',encoding='utf-8')
df
```

```python
df = df.loc[20211031:,'기준일ID':'행정동코드']
df
```

```python
df.columns = ['시간대구분', '행정동코드', '총생활인구수']
df
```

```python
code = df.loc[df['시간대구분'] == 0, '행정동코드'].tolist()
code
```

```python
short = []
for i in code :
    temp = df.loc[df['행정동코드'] == i]
    short.append(temp['총생활인구수'].mean())
short
```

```python
short_dic = {}
for i in code :
    temp = df.loc[df['행정동코드'] == i]
    short_dic[i] = temp['총생활인구수'].mean()
short_dic
```

```python
df2 = pd.read_csv('행정동별 서울생활인구(장기체류 외국인).csv',encoding='utf-8')
df2
```

```python
df2 = df2.loc[20211031:,'기준일ID':'행정동코드']
df2
```

```python
df2.columns = ['시간대구분', '행정동코드', '총생활인구수']
df2
```

```python
code2 = df2.loc[df2['시간대구분'] == 0, '행정동코드'].tolist()
code2
```

```python
long = []
for i in code2 :
    temp = df2.loc[df2['행정동코드'] == i]
    long.append(temp['총생활인구수'].mean())
long
```

```python
long_dic = {}
for i in code2 :
    temp = df2.loc[df2['행정동코드'] == i]
    long_dic[i] = temp['총생활인구수'].mean()
long_dic
```

```python
len(long_dic)
```

```python
non_code = []
for i in code2 :
    if i not in code :
        non_code.append(i)
non_code
```

```python
final_dict = {}
for i in long_dic.keys() :
    if i in short_dic.keys() :
        final_dict[i] = short_dic[i]
    else :
        final_dict[i] = 0
final_dict   
```

```python
final = {}
for i in code2 :
    s = final_dict[i] + long_dic[i]
    final[i] = s
final
```

```python
df =pd.DataFrame(list(final.items()), columns=['행정동코드', '총생활인구수'])
df
```
