```python
import pandas as pd
df = pd.read_csv('서울특별시 응급실 위치 정보.CSV', encoding='cp949')
df = df[['주소', '행정동']]
df
```

```python
df['구'] = df.주소.str.split(' ').str[1]
df = df.drop(['주소'], axis=1)
df
```

```python
df_sort = df.sort_values(by=['구', '행정동'])
df_sort
```

```python
result = df_sort.groupby(by=['구', '행정동']).size()
result
```

```python
result.to_csv('서울시 행정동별 응급실 개수.csv', encoding='cp949')
```
