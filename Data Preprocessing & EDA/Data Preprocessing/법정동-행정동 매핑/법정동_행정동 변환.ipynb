```python
import pandas as pd
import numpy as np
import re
import random
```

```python
#수정 필요
#법정동을 행정동으로 바꿀 데이터 경로
#법정동에 해당하는 열의 이름은 '법정동'으로 되어 있어야함
data='/content/drive/Shareddrives/2021 팜 경진대회/코드/법정동-행정동 매핑/서울특별시 행정동별 응급실 위치 정보.xlsx'

#행정동으로 바꾼 데이터 이름과 경로 (ex:새마을금고_행정동교체)
file='/content/drive/Shareddrives/2021 팜 경진대회/코드/법정동-행정동 매핑/서울특별시 행정동별 응급실 위치 정보.xlsx' 
```

```python
def mappingBeobIntoHaeng(data,file):
  compare='/content/drive/Shareddrives/2021 팜 경진대회/코드/법정동-행정동 매핑/compare_dong.xlsx'
  map='/content/drive/Shareddrives/2021 팜 경진대회/데이터/서울_행정동_법정동_매핑.csv'
  compare_data=pd.read_excel(compare)
  df_map=pd.read_csv(map)
  df_data=pd.read_excel(data)

  real_beobjeong=df_map['lgdng_nm'].unique().tolist()
  real_haengjeong=df_map['adstrd_nm'].unique().tolist()

  #법정동 인덱스 뽑아오기
  data_beobjeong=df_data['법정동'].tolist()
  data_beobjeong_match=[]
  for i in data_beobjeong:
    if i in real_beobjeong:
      # print(i,"에 해당하는 행정동들은 ",df_map.index[df_map['lgdng_nm']==i].tolist())
      data_beobjeong_match.append(df_map.index[df_map['lgdng_nm']==i].tolist())
    elif i in real_haengjeong and i not in real_beobjeong:
      data_beobjeong_match.append(i)
    else:
      data_beobjeong_match.append(None)

  #인덱스에 따른 행정동 추출 과정
  data_haengjeong=[]
  for i in data_beobjeong_match:
    # print("-------매치과정-------")
    # print("대상:",i)
    if isinstance(i,list):
      if len(i)==1:
        # print(i,"는",df_map['adstrd_nm'].iloc[i[0]],"으로 매치")
        data_haengjeong.append(df_map['adstrd_nm'].iloc[i[0]])
      elif len(i)>1:
        rand_num=random.choice(i)
        # print(i,"에서 ",rand_num,"추출")
        # print(rand_num,"는",df_map['adstrd_nm'].iloc[rand_num],"으로 매치")
        data_haengjeong.append(df_map['adstrd_nm'].iloc[rand_num])
    else:
      # print(i,"는 ",i)
      data_haengjeong.append(i)
  #   print("---------------------")
  # print("-----------------------------------------------------------")

  #'행정동_법정동_매핑.csv'와 '산일구비 데이터셋' 행정동명 일치 과정
  real_haeng=compare_data['실제 행정동'].tolist()
  san1_haeng=compare_data.iloc[:,2].tolist()
  for i,v in enumerate(data_haengjeong):
    if v in real_haeng:
      # print("대상:",v)
      id=real_haeng.index(v)
      temp=san1_haeng[id]
      data_haengjeong[i]=temp
      # print("결과:",data_haengjeong[i])

  df_data['행정동']=data_haengjeong
  df_data.to_excel(file,index=False)
  print("저장완료 :) !!")
```

```python
mappingBeobIntoHaeng(data,file)
```
