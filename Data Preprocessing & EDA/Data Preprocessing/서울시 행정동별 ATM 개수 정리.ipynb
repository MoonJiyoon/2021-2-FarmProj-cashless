```python
import pandas as pd
df = pd.read_csv('주요은행5개 통합 atm.CSV', encoding='cp949')
df = df[['도로명주소', '행정동']]
df
```

```python
df['구'] = df.도로명주소.str.split(' ').str[1]
df = df.drop(['도로명주소'], axis=1)
df
```

```python
df_sort = df.sort_values(by=['구', '행정동'])
```

```python
result = df_sort.groupby(['구', '행정동']).size()
result
```

```python
result.to_csv('서울시 행정동별 ATM 개수.csv', encoding='cp949')
```
