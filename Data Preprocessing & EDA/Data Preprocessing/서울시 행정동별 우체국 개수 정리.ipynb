```python
import pandas as pd
df = pd.read_csv('서울시 우체국 정보 (동별) 통계.csv', encoding='cp949')
```

```python
df = df.drop(['키값', '기관코드', '유형1', '유형2', '대표기관명', '전체기관명', '대표전화번호', '새우편번호', '주소'], axis=1)
```

```python
df_sort = df.sort_values(by=['행정구', '행정동'])
```

```python
df = df_sort.reset_index(drop=True)
df
```
