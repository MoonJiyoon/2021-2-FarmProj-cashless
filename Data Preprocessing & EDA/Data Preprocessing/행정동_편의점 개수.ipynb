```python
df1 = pd.read_excel('편의점 수.xlsx', sheet_name = 'Sheet1')
df1
```

```python
df1.info()
```

```python
df1.isna().sum()
```

```python
df1 = df1.dropna()
df1
```

```python
df1.reset_index(drop=True, inplace=True)
df1
```

```python
dong = []
num = []
for i in df1['행정구역'] :
    dong.append(i)
for j in df1['전체 점포수'] :
    num.append(j)
```

```python
# 데이터셋에는 있지만 편의점에는 없는 것
for i in d :
    if i not in dong :
        print(i)
```

```python
# 편의점에는 있지만 데이터셋에는 없는 것
for i in dong :
    if i not in d :
        print(i)
```

```python
dict = {dong[i]:num[i] for i in range(len(num))}
dict
```

```python
temp = []
for i in d :
    for j in dong :
        if i == j :
            temp.append(dict[i])

    if i == '항동' :
        temp.append(8)
    if i == '상일1동' :
        temp.append(7)  
    if i == '상일2동' :
        temp.append(0) 

temp
```

```python
df['편의점 수'] = temp
```

```python
df.to_csv('행정동별 편의점 수.csv')
```
